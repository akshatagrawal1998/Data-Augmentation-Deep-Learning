# Data Augmentation

It is a technique by which we generate more data from the data we already have by applying various transformation techniques like 
flipping, rotation, horizontal or vertical shifting and so on

Pros:
    - Saves cost if we have less data
    - It helps to reduce overfitting
    

- https://blog.keras.io/building-powerful-image-classification-models-using-very-little-data.html

```python
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
```

```python
from keras.utils import load_img, img_to_array
```

```python
img = load_img(r'C:\Users\HP\Deep Learning\Data Augmentation\SRK image.jpg', target_size=(300,200))  # this is a PIL image
```

```python
img
```

```python
type(img) # In OPEN CV it's like numpy array but here it is PIL(Python Imaging Library) type
```

```python
x = img_to_array(img)  # this will convert img(PIL) to a Numpy array with shape (3, 150, 150)
```

```python
x.shape
```

```python
type(x)
```

```python
# transformaions to apply on the given image

datagen = ImageDataGenerator(
        rotation_range=40, #angle by which image is to be rotated
        width_shift_range=0.2, # shift the image
        height_shift_range=0.2, # shift the image
        shear_range=0.2, # distort image
        zoom_range=0.2,  # zoom in and zoom out
        horizontal_flip=True
        )

# we can apply more transformations as needed
```

```python
input_batch =x.reshape(1,300,200,3) # to tell the system that we are giving 1 image to transform
```

```python
# the .flow() command below generates batches of randomly transformed images and saves the results to the `augmentation output/` directory

# there is another function in datagen class called "flow from directory" which is used when 
# we apply transformations on multiple images from a directory


i = 0

for batch in datagen.flow(input_batch, batch_size=1,
                          save_to_dir='augmentation output/', save_prefix='srk_transformed', save_format='jpeg'):
    i += 1
    if i > 20:
        break  # otherwise the generator would loop indefinitely.
```

```python
# now if we go to augmentation output folder, it will have 20 transformed images
```
